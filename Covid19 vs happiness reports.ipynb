```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
corona_dataset_csv=pd.read_csv('covid19_Confirmed_dataset.csv')
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
corona_dataset_csv.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
corona_dataset_csv.head(10)
```

```python
corona_dataset_aggregated=corona_dataset_csv.groupby('Country/Region').sum()
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.loc['India'].plot(label='India')
corona_dataset_aggregated.loc['China'].plot(label='China')
plt.legend()
```

```python
corona_dataset_aggregated.loc['China'][0:3].plot(label='China')
plt.legend()
```

```python
corona_dataset_aggregated.loc['India'].diff().plot()
```

```python
corona_dataset_aggregated.loc['India'].diff().max()
```

```python
countries=list(corona_dataset_aggregated.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
```

```python
max_infection_rates
```

```python
corona_dataset_aggregated['max_infection_rate']=max_infection_rates
```

```python
corona_dataset_aggregated.head(4)
```

```python
corona_data=pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
```

```python
corona_data.head(4)
```

```python
happiness_report_csv=pd.read_csv('worldwide_happiness_report.csv')
```

```python
happiness_report_csv.head(4)
```

```python
useless_columns=['Overall rank', 'Score', 'Generosity', 'Perceptions of corruption']
```

```python
happiness_report_csv.drop(useless_columns, axis=1, inplace=True)
```

```python
happiness_report_csv.head(4)
```

```python
happiness_report_csv.set_index("Country or region", inplace=True)
```

```python
happiness_report_csv.head(4)
```

```python
happiness_report_csv.shape
```

```python
data=corona_data.join(happiness_report_csv, how="inner")
```

```python
data.head(4)
```

```python
data.corr()
```
